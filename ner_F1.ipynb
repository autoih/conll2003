```python
import requests

url = 'http://0.0.0.0:5000/'
model_endpoint =  'model/predict'
complete_url = url + model_endpoint

cnt_err=0
cnt = 0
pred_tags = []
input_terms = []
with open('input_data_3.txt') as f:
    lines = f.readlines()
    for line in lines:
        line = line.lower()
        try:
            file_form = {"text":line}
            # Post the image to the rest API using the requests library
            r = requests.post(url=complete_url, json=file_form)
            # Return the JSON
            response = r.json()
#             print('==========')
#             print(line)
#             print(response.items())
#             print('+++++++++++++')
            pred_tags.append(response['prediction']['tags'])
            input_terms.append(response['prediction']['terms'])
            cnt += 1
        except:
            print('Error')
            cnt_err += 1

with open('output_pred_tags.txt', 'w') as f:
    f.write("%s\n" % pred_tags)    
    
with open('output_input_terms.txt', 'w') as f:
    f.write("%s\n" % input_terms)  

print(cnt)
print(cnt_err)
print('FINISHED..................')
```
